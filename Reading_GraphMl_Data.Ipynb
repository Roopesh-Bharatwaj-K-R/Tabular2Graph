<b> Reading GraphMl Data Using Networkx in Python</b>

```python
import networkx as nx
import matplotlib.pyplot as plt
# Specify the path to your GraphML file
file_path = "rwh.graphml"
# Read the GraphML file
graph = nx.read_graphml(file_path)
# Visualize the graph
plt.figure(figsize=(20, 13))  # Set the size of the figure
nx.draw(graph, with_labels=True)  # Draw the graph with labels
plt.show()  # Display the graph
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
# Step 2: Preprocess the graph data
adj_matrix = nx.adjacency_matrix(graph)
adj_matrix = torch.FloatTensor(adj_matrix.toarray())
```
